```python
# Importing:
import numpy as np
import pandas as pd
from IPython.display import clear_output as co
import matplotlib.pyplot as plt
import tensorflow as tf
from six.moves import urllib
import inspect
import tensorflow.compat.v2.feature_column as fc

if not hasattr(inspect, 'getargspec'):
    inspect.getargspec = inspect.getfullargspec
```

```python
# Creating constants for supproting later
COLUMN_NAMES = ['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth', 'Species']
SPECIES = ['Setosa', 'Versicolor', 'Virginica']
```

```python
# Installing datasets paths
trainPath = tf.keras.utils.get_file("iris_training.csv", "https://storage.googleapis.com/download.tensorflow.org/data/iris_training.csv")
testPath = tf.keras.utils.get_file("iris_test.csv", "https://storage.googleapis.com/download.tensorflow.org/data/iris_test.csv")

# Creating Pandas DataFrame
train_df = pd.read_csv(trainPath, names=COLUMN_NAMES, header=0) # Creating the DataFrame with the columns names defined earlier. The parameter  
                                                                # header = 0 means that the previous parameter should be at the first line #
                                                                # (header)

test_df = pd.read_csv(testPath, names=COLUMN_NAMES, header=0)
```

```python
# Create the labels (y-axis)
train_y = train_df.pop('Species')
test_y = test_df.pop('Species')

COLUMN_NAMES.remove("Species") # This column has been moved to others vars
```

# More about the dataframe:
train_df: 120 entries with 4 features
test_df: 29 entries with 4 features

```python
# Creating the Input Function
def input_fun(features, labels, training=True, batch_size=256):
    # Converting the inputs to a DataSet
    dataset = tf.data.Dataset.from_tensor_slices((dict(features), labels))

    # Shuffle and reapeat if you are in training mode.
    if training:
        dataset = dataset.shuffle(1000).repeat()
    
    return dataset.batch(batch_size)

# Observe: We did not use Epochs.
```

```python
# Defining the features column
FEATURES_COLUMN = []
for key in train_df.keys():
    FEATURES_COLUMN.append(tf.feature_column.numeric_column(key=key))
```

# Building the model
Since we are working with classification, there is a variety of different models we can use, such as:

DNNClassifier (Deep Neural Network)
Linear Classifier

I am going to use de DNN because there may be non-linear relations in our DatSet.

```python
classifier = tf.estimator.DNNClassifier(
    feature_columns=FEATURES_COLUMN, # The features columns of our DataSet
    hidden_units=[30,10], # 2 Hidden Layers of 30 and 10 nodes (Neurons), for 3 Hidden Layers should be something: [amountNodes ,          
                        # amountNodes2, amountNodes3]
    n_classes=3 # The amount of clases the model will choose between
)
```

```python
#Training!
classifier.train(
    input_fn= lambda: input_fun(train_df, train_y, training=True), # Did this because we did not defind an inner function at input_fun
    steps=5000 # The times the model will update the weights and biases afet a batch
)
```

```python
# Checking output:
result = classifier.evaluate(input_fn=lambda: input_fun(test_df, test_y, training=False))
print("\033[32mAccuracy: " + str(result['accuracy']*100) + "%\033[0m")
```

```python
# Predicting according to system.input:
def input_pred_fn(features, batch_size=256):
    # Convert the inputs to a Dataset without labels.
    return tf.data.Dataset.from_tensor_slices(dict(features)).batch(batch_size)

predict = {}

for feature in COLUMN_NAMES:
    value = input(feature + ": ")
    predict[feature] = [float(value)]

predictions = classifier.predict(input_fn=lambda: input_pred_fn(predict))
for pred_dict in predictions:
    class_id = pred_dict['class_ids'][0]
    probability = pred_dict['probabilities'][class_id]

    print('\033[32mPrediction is "{}" ({:.1f}%)\033[32m'.format(
        SPECIES[class_id], 100 * probability))

```
